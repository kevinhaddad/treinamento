### 21. Crie um programa em Python que receba uma string como entrada e imprima a string sem as vogais.

```python
string = input("Digite uma frase para suprimir as vogais: ")

novaString = ''
vogal = "aeiouAEIOU"
print ("Você escreveu: " + string)

for letra in string:
  if letra in vogal:
    novaString += ""
  else:
    novaString += letra
    
print(novaString)
```

### 22. Crie um programa em Python que imprima a soma dos n primeiros números primos, onde n é um número fornecido pelo usuário.

```python
n = int(input("Digite um número: "))

soma = 0
count = 0

for num in range(2, 100):
    for i in range(2, num):
        if (num % i) == 0:
            break
    else:
        soma += num
        count += 1
        if count == n:
            break
    if count == n:
        break

print("A soma dos primeiros", n, "números primos é:", soma)

      
```

### 23. Crie um programa em Python que receba uma lista de números e imprima a soma dos números pares e a soma dos números ímpares presentes na lista.

```python
numeros = input("Digite uma lista de números separados por espaço: ").split()
numeros = [int(x) for x in numeros]

somaPar = 0
somaImpar = 0

for i in numeros:
  if (i % 2 == 0):
    somaPar += i
  else:
    somaImpar += i
    
print("A soma dos pares é: " + str(somaPar) + " e a soma dos ímpares é: " + str(somaImpar))

```

### 24. Crie um programa em Python que receba uma lista de strings e retorne uma lista com todas as strings que começam e terminam com a mesma letra.

```python
palavras = ["amor", "futebol", "verdade", "banana", "arara", "casa", "sol"]
palavras_selecionadas = []

 
for i in palavras:
  if i[0] == i[-1]:
    palavras_selecionadas.append(i)

print(palavras_selecionadas)
```
