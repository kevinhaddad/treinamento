### 1. Criar um array com 10 números inteiros aleatórios entre 1 e 50 e calcular a média e o desvio padrão

```
import numpy as np

arr = np.random.randint(1,51, size=10)

media = np.mean(arr)

desvio_padrao = np.std(arr)

print(arr)
print(media)
print(desvio_padrao)
```

### 2. Criar um array de 10 números aleatórios e calcular a raiz quadrada de todos os valores menores ou iguais a 0,5

```
arr = np.random.rand(10)

arr[arr <= 0.5] = np.sqrt(arr[arr <= 0.5])

print(arr)
```

### 3. Criar uma matriz 3D de tamanho 3x3x3 com valores aleatórios e exibir o valor máximo de cada linha de cada matriz 2D dentro dessa matriz 3D

```
matriz = np.random.rand(3, 3, 3)

valor_maximo = np.amax(matriz, axis=1)

print("Matriz 3d:\n", matriz)
print("Valores máximos de cada linha de cada matriz 2d:\n", valor_maximo)
```

### 4. Criar uma matriz de tamanho 5x5 com valores aleatórios e subtrair a média de cada linha dessa matriz

```
matriz = np.random.rand(5,5)

print(matriz)

for i in range(5):
  media = np.mean(matriz[i])
  matriz[i] -= media
  
print(matriz)
```

### 5. Criar uma matriz de tamanho 4x4 com valores aleatórios e calcular a média e o desvio padrão das diagonais da matriz

```
matriz = np.random.rand(4,4)

print(matriz)
#calculo da media e desvio padrao das diagonais da matriz

diagonalPrincipal = np.diag(matriz) #pega a diagonal da matriz
diagonalSecundaria = np.diag(np.fliplr(matriz))#o fliplr inverte a matriz e assim conseguimnos pegar a diagonal secundaria da matriz principal uma vez que ela foi invertida
mediaDiagonal = np.mean([diagonalPrincipal, diagonalSecundaria]) #faz a média das diagonais que foram extraídas da matriz principal
desvioPadrao = np.std([diagonalPrincipal, diagonalSecundaria]) #calcula o desvio padrao das duas diagonais

print(f'Diagonal Principal: {diagonalPrincipal}')
print(f'Diagonal Secundaria: {diagonalSecundaria}')
print(f'Media diagonais: {mediaDiagonal}')
print(f'Desvio Padrão: {desvioPadrao}')
```
